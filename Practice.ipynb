```python
# Pyspark initialization and session creation

import findspark
findspark.init()

from pyspark.mllib.feature import Word2Vec
from pyspark import SparkConf
from pyspark.sql import SparkSession

conf = SparkConf()

spark = SparkSession \
    .builder \
    .appName('Word2Vec Implementation') \
    .getOrCreate()

spark.conf.set('spark.sql.caseSensitive', False)
```

```python
'''
1. Process the doc to remove the punctuation marks and tokenize the words
2. Join the tokenized words to form sentences and use remove_stopwords method from gensim to remove the stopwords such as
    'the', 'with', 'of', etc
    Source: https://stackabuse.com/removing-stop-words-from-strings-in-python/#usingpythonsgensimlibrary
3. DocTokens obtained is of the format:    
            docTokens =  ['sentence 1', 'sentence 2',...]
'''
for doc in docs:
    processedText = gensim.utils.simple_preprocess(doc)
    removedStopWords = remove_stopwords(' '.join(processedText))
    docTokens.append(removedStopWords)

docTokensCollection = spark.sparkContext.parallelize(docTokens).map(lambda line: line.split(" "))
```

```python
# Import libraries
import requests
from bs4 import BeautifulSoup

# URL from which pdfs to be downloaded
url = "https://www.geeksforgeeks.org/how-to-extract-pdf-tables-in-python/"

# Requests URL and get response object
response = requests.get(url)

# Parse text obtained
soup = BeautifulSoup(response.text, 'html.parser')

# Find all hyperlinks present on webpage
links = soup.find_all('a')

i = 0

# From all links check for pdf link and
# if present download file
for link in links:
    if ('.pdf' in link.get('href', [])):
        i += 1
        print("Downloading file: ", i)

        # Get response object for link
        response = requests.get(link.get('href'))

        # Write content in pdf file
        pdf = open("pdf"+str(i)+".pdf", 'wb')
        pdf.write(response.content)
        pdf.close()
        print("File ", i, " downloaded")

print("All PDF files downloaded")
```
